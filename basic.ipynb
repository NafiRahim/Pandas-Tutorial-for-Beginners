Some time if your data is large its hard to verify in bare eye that is there any null value or empty value in the data. You can use the isnull() method to check if there is any null value in the data. The output is a boolean value indicating whether the value that is passed into the argument is in fact missing data.

You can also use the notnull() method to check if there is any non-null value in the data. The output is a boolean value indicating whether the value that is passed into the argument is in fact not missing data.

```python
df.isnull() # print number of null values in each column
```

```python
df.notnull() # print number of not null values in each column
```

But only isnull or is not null function not that usefull in a big data set. We need to know how many null value in a column or row. We can use the sum() method to count the number of True values.

```python
# so to know the total number of null values in each column we can use sum() function with isnull() function

df.isnull().sum() # print total number of null values in each column
```

```python
df.notnull().sum() # print total number of not null values in each column
```

### if we like to get the specific row of the dataframe, we can use loc[] function. loc[] function takes two argument, first one is the index of the row and second one is the column name.

But if just pass the index of the row, it will return all the columns of that row.

```python
df.loc[0] # print all data of first row
```

#### if we want to get from this row to that row information, we can use slicing.

```python
# slicing data
df.loc[0:3] # print all data of first 4 rows
```

```python
# slicing data with specific column
df.loc[0:3, ["Name", "Age"]] # print all data of first 4 rows
```

### Another way to get the specific row of the dataframe, we can use iloc[] function. iloc[] function takes two argument, first one is the index of the row and second one is the column index.


```python
df.iloc[0:3, 0:2] # print all data of first 4 rows and first 2 columns
```

### The main difference between iloc and loc is that, loc function takes column name as argument but iloc function takes column index as argument.
As a user i think loc function is more useful than iloc function. because we can easily understand the column name than column index.
In a large dataset, it would be difficult to understand the column index. So, I think loc function is more useful than iloc function.

### Now lets do some real work, such  as make few changes in the data, create new table from the existing table, create new column from the existing column etc.

Lets say, as a teacher i want to give 5 marks extra to the student who got more than 85 marks in math. We can do that by using loc function.

```python
# give 5 marks extra to each student who got more than 85 marks in Maths but less than 95 marks in Maths

df.loc[(df.Maths > 85) & (df.Maths < 95), "Maths"] = df.Maths + 5
```

### Now i want to create new Dataframe for each section. We can do that by using loc[] function.

```python
# make seperate df for each section

Section_A = df.loc[df.Section == "A"]
Section_B = df.loc[df.Section == "B"]
Section_C = df.loc[df.Section == "C"]
```

Since we have created new dataframe for each section, and also drop Age columns we don't need the section column anymore. We can drop the section column by using drop() function.

```python
# drop section column from each df

Section_A = Section_A.drop("Section", axis=1)
Section_B = Section_B.drop("Section", axis=1)
Section_C = Section_C.drop("Section", axis=1)
```

### Lets sum of the marks of each section. And put the total in a new column.

```python
Section_A["Total"] = Section_A.Maths + Section_A.English + Section_A.Science + Section_A.History
Section_B["Total"] = Section_B.Maths + Section_B.English + Section_B.Science + Section_B.History
Section_C["Total"] = Section_C.Maths + Section_C.English + Section_C.Science + Section_C.History
```

```python
Section_A.describe()
```

Lets make Age columns into a string, so that statistics summary will not show the Age column.

```python
# convert age.int to string
Section_A.Age = Section_A.Age.astype(str)
Section_B.Age = Section_B.Age.astype(str)
Section_C.Age = Section_C.Age.astype(str)
```

```python
Section_A.describe()
```

### Now, we can export the data into csv file by using to_csv() function.

```python
Section_A.to_csv("Section_A.csv", index=False)
Section_B.to_csv("Section_B.csv", index=False)
Section_C.to_csv("Section_C.csv", index=False)

# index=False will not save index column in csv file
```
